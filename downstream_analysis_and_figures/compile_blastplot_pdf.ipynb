```python
import os
from fpdf import FPDF
import pandas as pd
import ast
```

```python
td=set([x.split(".tsv")[0] for x in os.listdir("data_upload_3_24_2025/blast_round_one_data")])&set(os.listdir("blast_result_plots"))
```

```python
df=pd.read_csv("merged_chimera_clusters.csv",index_col=1)
cmap={}
for index, row in df.iterrows():
    cmap[row.cluster_id]=ast.literal_eval(row.merged_chimera_clusters)
 
    
```

```python
## add sequences removed via orthologous clustering
cmap[1]=['XP_034487048.1','XP_031781467.1']
cmap[7]=['XP_046456339.1', 'XP_046645476.1', 'XP_046456333.1', 'XP_046456336.1']
```

```python

# Path to the directory containing SVG files, with round1 and round2 svgs in the same subdirectories per svg
input_directory = 'blast_result_plots'

# Output PDF file
pdf_output = "blastplots_merged.pdf"

# Create a PDF document
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)

# Iterate through all clusters, then all sequences with blastplots in the cluster
for c in cmap:
    for index in cmap[c]:
        dir_name=index
        dir_path = os.path.join(input_directory, dir_name)

        # Ensure we are processing subdirectories (not files)
        if os.path.isdir(dir_path):
            # Create a new page for the PDF
            pdf.add_page()
            pdf.set_font("Arial", size=16)
            if dir_name in df.index:
                pdf.cell(200, 10, txt=dir_name+f" (cluster {c}, representative sequence)", ln=True, align='C')
            else:
                pdf.cell(200, 10, txt=dir_name+f" (cluster {c})", ln=True, align='C')
            pdf.cell(200, 10, txt='Blast Round 1', ln=True, align='C')
            svg_path = os.path.join(dir_path, 'round1_plot.svg')
            pdf.image(svg_path, x=pdf.l_margin, y=pdf.t_margin + 20, w=190.00155, h=76.874)
            y_position_second_plot = pdf.get_y() + 80
            pdf.cell(200, y_position_second_plot+60, txt='Blast Round 2', ln=True, align='C')
            svg_path = os.path.join(dir_path, 'round2_plot.svg')
            pdf.image(svg_path, x=pdf.l_margin, y=y_position_second_plot + 10, w=190.00155, h=76.874)


pdf.output(pdf_output)

print(f"PDF generated successfully: {pdf_output}")
```
